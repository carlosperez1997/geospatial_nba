```python
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
file_path = "./plays.json"

with open(file_path, 'r') as file:
    plays = json.load(file)

plays_df = pd.DataFrame(plays['plays'])
plays_df.head(10)

plays_df['x'] = plays_df['coordinate'].apply(lambda x: x['x'])
plays_df['y'] = plays_df['coordinate'].apply(lambda x: x['y'])
plays_df
```

```python
file_path = "./scores.json"

# Open the JSON file and load its contents
with open(file_path, 'r') as file:
    scores = json.load(file)

scores_df = pd.DataFrame(scores["plys"])
scores_df.head(20)
```

```python
all_data = pd.merge(plays_df, scores_df, how='left', on='id')
all_data.head(10)
all_data.to_csv('plays.csv')
```
